```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
csv_data = pd.read_csv('all_data.csv')
csv_data = csv_data.sort_values(by='Year')
display(csv_data)
print(csv_data.info())
print(csv_data.describe())
```

```python
country = csv_data.Country
year = csv_data.Year
life_exp = csv_data['Life expectancy at birth (years)']
gdp = csv_data.GDP

sns.boxplot(x = 'Year', y = 'Life expectancy at birth (years)', palette = "pastel", data = csv_data)
plt.title("Expectativa de vida desde o ano 2000 atÃ© 2015 nos 6 paÃ­ses")
plt.xlabel("Anos")
plt.ylabel("Expectativa de vida (anos)")
plt.xticks(rotation=30)
plt.show()
plt.clf()

plt.bar(year, life_exp)
plt.title("Expectativa de vida desde o ano 2000 atÃ© 2015 nos 6 paÃ­ses")
plt.xlabel("Anos")
plt.ylabel("Expectativa de vida (anos)")
plt.show()
plt.clf()


```

```python
sns.boxplot(x = 'Year', y = 'GDP', palette = "pastel", data = csv_data)
plt.title("Produto Interno Bruto dos 6 paÃ­ses do ano 2000 a 2015")
plt.xlabel("Anos")
plt.ylabel("Produto Interno Bruto ($)")
plt.xticks(rotation=30)
plt.show()
plt.clf()

plt.bar(year, gdp)
plt.title("Expectativa de vida dos 6 paÃ­ses do ano 2000 atÃ© 2015")
plt.xlabel("Anos")
plt.ylabel("Produto Interno Bruto ($)")
plt.show()
plt.clf()
```

```python
sns.boxplot(x = 'Country', y = 'Life expectancy at birth (years)', palette = "pastel", data = csv_data)
plt.xlabel("Anos")
plt.ylabel("Expectativa de vida (anos)")
plt.title("Expectativa de vida por paÃ­s")
plt.xticks(rotation=30)
plt.show()
plt.clf()

sns.boxplot(x = 'Country', y = 'GDP', palette = "pastel", data = csv_data)
plt.xlabel("Anos")
plt.ylabel("Produto Interno Bruto ($)")
plt.title("Produto Interno Bruto por paÃ­s")
plt.xticks(rotation=30)
plt.show()
plt.clf()
```

```python
countries_strings = csv_data.Country.unique()

for country in countries_strings:
    data_per_country = csv_data[csv_data.Country == country]
    plt.plot(data_per_country['Year'], data_per_country['Life expectancy at birth (years)'], label = country)
    plt.legend(fontsize='small')
    plt.title("EvoluÃ§Ã£o da expectativa de vida por paÃ­s")
    plt.xlabel("Anos")
    plt.ylabel("Expectativa de vida (anos)")
plt.show()
plt.clf()

for country in countries_strings:
    data_per_country = csv_data[csv_data.Country == country]
    plt.plot(data_per_country['Year'], data_per_country['GDP'], label = country)
    plt.legend(fontsize='small')
    plt.title("EvoluÃ§Ã£o do PIB por paÃ­s")
    plt.xlabel("Anos")
    plt.ylabel("Produto Interno Bruto ($)")
plt.show()
plt.clf()
```

```python
# plt.scatter(gdp, life_exp)

for country in countries_strings:
    data_per_country = csv_data[csv_data.Country == country]
    sns.lmplot(x = "Life expectancy at birth (years)", y = "GDP", data = data_per_country, line_kws={'color': 'black'}, lowess=True, scatter_kws={"s": 10})
    plt.title("CorrelaÃ§Ã£o entre o PIB e a expectativa de vida no " + country)
    plt.xlabel("Expectativa de vida (anos)")
    plt.ylabel("Produto Interno Bruto ($)")
    plt.show()
    plt.clf()
```

```python
plt.hist(life_exp)
plt.title("Histograma da expectativa de vida dos 6 paÃ­ses")
plt.xlabel("Expectativa de vida (anos)")
plt.ylabel("FrequÃªncia")
plt.show()
plt.clf()

plt.hist(gdp)
plt.title("Histograma do produto interno bruto dos 6 paÃ­ses")
plt.xlabel("Produto Interno Bruto ($)")
plt.ylabel("FrequÃªncia")
plt.show()
plt.clf()
```
